## 2-Data Structure

The PElectricity1344 dataset contains **wholesale electricity prices** from the New England electricity market.
These are  weekly series of electricity price each hour of each day during $678 weeks$ in eight regions of New England (NE),
| Dimension | Values | Description |
|-----------|--------|-------------|
| Days | 7 | Thursday through Wednesday |
| Hours | 24 | 1 AM CT through 12 AM |
| Regions | 8 | New England pricing zones |
| **Total Series** | **1344** | 7 × 24 × 8 = 1344 |

Data is described in the book at page 221

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read 
df = pd.read_csv('../PElectricity1344.csv')
df
```

```python
# Choisir UNE colonne (ex : la première vraie série)
serie = df.iloc[:, 0]
# Tracer
plt.figure(figsize=(10,4))
plt.plot(serie)
plt.xlabel("Weeks")
plt.ylabel("Electricity price")
plt.title("Price Growth – A serie  (day/hour/region)")
plt.show()
```
